```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataacc = pd.read_csv("dataacc Esp2 .csv")
dataacc = dataacc.drop(['Unnamed: 0'], axis=1)
dataaccbagi3 = pd.read_csv("dataaccbagi3 Esp2 .csv")
dataaccbagi3 = dataaccbagi3.drop(['Unnamed: 0'], axis=1)
amplitudemaks = pd.read_csv("amplitudeintervalxdf Esp2 .csv")
amplitudemaks = amplitudemaks.drop(['Unnamed: 0'], axis=1)
data_size = 5000
```

```python
#DataPlot Acc
n=10

x = [t//n for t in range (1,data_size) ]
arr = dataacc.to_numpy()
x_accel1 = [arr[i][0] for i in range(len(arr)-1)]
ESP1_RSSI = [arr[i][1] for i in range(len(arr)-1)]

plt.subplot(2, 1, 1)
plt.plot(x, x_accel1, color='r',linewidth = 1)
plt.title('Plot Accelerometer X & RSSI',fontsize = 10)
plt.ylabel('Xesp2',fontsize = 10)

plt.subplot(2, 1, 2)
plt.plot(x, ESP1_RSSI , color='r',linewidth = 1)
plt.xlabel('time',fontsize = 10)
plt.ylabel('RSSIESP2 (dBm)',fontsize = 10)

plt.show()
```

```python
#dataplot accbagi3
n=10

x = [t//n for t in range (1,data_size) ]
arr = dataaccbagi3.to_numpy()
x_accel1 = [arr[i][0] for i in range(len(arr)-1)]
ESP2_RSSI = [arr[i][1] for i in range(len(arr)-1)]

plt.subplot(2, 1, 1)
plt.plot(x, x_accel1, color='g',linewidth = 1)
plt.title('Plot Accelerometer Xbagi3 & RSSI',fontsize = 10)
plt.ylabel('Xesp2',fontsize = 10)

plt.subplot(2, 1, 2)
plt.plot(x, ESP2_RSSI , color='g',linewidth = 1)
plt.xlabel('time ',fontsize = 10)
plt.ylabel('RSSIESP2 (dBm)',fontsize = 10)

plt.show()
```

```python
#Acc vs Acc/3
size = 20

x = [t for t in range (1,(data_size//size)) ]
arr = dataacc.to_numpy()
x_accel = [arr[i][0] for i in range((len(arr)-1)//size)]

arrbagi3 = dataaccbagi3.to_numpy()
x_accelbagi3 = [arrbagi3[i][0] for i in range((len(arrbagi3)-1)//size)]

plt.subplot(1, 1, 1)
plt.plot(x, x_accel, color='r',linewidth = 2)
plt.plot(x, x_accelbagi3, color='g',linewidth = 2)
plt.title('Plot Accelerometer X vs X/3', fontsize = 10)
plt.xlabel('time ' , fontsize = 10)
plt.ylabel('X Bagi3 (H) & X (M)', fontsize = 10)

#plt.rcParams["figure.figsize"] = (120,100)
plt.show()
```

```python
#RSSI vs Acc/3 
size = 20

x = [t for t in range (1,(data_size//size)) ]
arr = dataaccbagi3.to_numpy()
x_rssi = [arr[i][1] for i in range((len(arr)-1)//size)]

arrbagi3 = dataaccbagi3.to_numpy()
x_accelbagi3 = [arrbagi3[i][0] for i in range((len(arrbagi3)-1)//size)]

plt.subplot(2, 1, 1)
plt.plot(x, x_accelbagi3, color='g',linewidth = 2)
plt.title('Plot Accelerometer X/3 VS RSSI', fontsize = 10)
plt.xlabel('time' , fontsize = 10)
plt.ylabel('X esp2', fontsize = 10)

plt.subplot(2, 1, 2)
plt.plot(x, x_rssi, color='r',linewidth = 2)
plt.xlabel('time' , fontsize = 10)
plt.ylabel('RSSI (dBm) esp2', fontsize = 10)

#plt.rcParams["figure.figsize"] = (120,100)
plt.show()
```
